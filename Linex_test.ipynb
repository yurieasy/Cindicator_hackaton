```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def linex(err, alpha):
    return np.exp(-alpha * err) + alpha * err - 1
```

```python
X = np.arange(-10, 10, 0.1)
Y = linex(X, -0.35)
```

```python
plt.plot(X,Y)
```
